```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import statsmodels.formula.api as smf
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
```

```python
##Extract dataset
df = pd.read_csv("https://raw.githubusercontent.com/HumayDS/Big-data-analysis/main/diabetes.csv")
df.drop('Unnamed: 0' , axis = 1, inplace = True)
df = df.dropna()
df.head(50)
```

```python
# Get Information of target variable
df["Outcome"].value_counts()
```

```python
#Determine x and
y = df["Outcome"]
X = df.drop(["Outcome"], axis=1)
```

```python
# prompt: create logistic regression y, X ith intercept
X = sm.add_constant(X)
logit_model = sm.Logit(y, X)
result = logit_model.fit()
```

```python
#Cretae Log regression with statsmodel (For interprete )
loj = sm.Logit(y, X)
loj_model= loj.fit()
loj_model.summary()
```

```python
# Create model with sklearn
from sklearn.linear_model import LogisticRegression
loj = LogisticRegression(solver = "liblinear")
loj_model = loj.fit(X,y)
loj_model
```

```python
#predict y
y_pred = loj_model.predict(X)
y_pred[0:10]
```

```python
y.head()
```

```python
##View predicted probability
y_probs = loj_model.predict_proba(X)
y_probs = y_probs[:,1]
y_probs[0:10]
```

```python
#View summary
i = np.arange(len(y_pred))
Data = pd.DataFrame({'y_real' : pd.Series(y, index=i),'y_pred' : pd.Series(y_pred, index = i), 'y_probs' : pd.Series(y_probs, index = i)})
Data.head(15)
```

#Let's calculate the Accuracy value (true  predictions / all predictions ---- 77% of our predictions are correct):
#(TP+TN)/(TP+TN+FP+FN)


```python
y_pred = [1 if i > 0.5 else 0 for i in y_probs]
```

```python
accuracy_score(y, y_pred)
```

#Precision --- What percentage of what we predict as diabetes is actually diabetes?
# TP/(TP+FP)

```python
print(precision_score(y,y_pred))
print(149/(149+54))
```

#Recall -  Sensitivity  -- True Pozitive rate --  - What percentage of people with diabetes were actually predicted to be diabetic?
#TP/(TP+FN)

```python
print(recall_score(y,y_pred))
print(149/(149+119))
```

# F1 Skoru --  2 * (Presicion * recall)/(Presicion + recall)
#Balanced presicion and recall


```python
f1_score(y,y_pred)
```
