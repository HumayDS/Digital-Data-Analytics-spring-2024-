```python
from warnings import filterwarnings
filterwarnings('ignore')
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/HumayDS/Big-data-analysis/main/Advertising.csv")
df.head()
```

```python
df.drop("Unnamed: 0" , axis =1 , inplace = True)
```

```python
df.isnull().sum()
```

```python
X = df[["TV"]]
X[0:5]
```

```python
X = sm.add_constant(X)
```

```python
y = df["sales"]
```

```python
lm = sm.OLS(y,X)
```

```python
model = lm.fit()
```

```python
model.summary()
```

```python
print("Sales = " +  str("%.2f" % model.params[0]) + " + TV" + "*" + str("%.2f" % model.params[1]))
```
