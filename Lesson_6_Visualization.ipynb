```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import dataset and view
df = pd.read_excel("https://raw.githubusercontent.com/HumayDS/Digital-Data-Analytics-spring-2024-/main/tips2.xlsx")
print(df.head())
print(df.info())
print(df.describe().T)
print(df.isnull().sum())
```

```python
# write dataset as excell format

df.to_excel('tips2.xlsx', index=False)
```

```python
##Delete column
df.drop("Unnamed: 0" , axis = 1, inplace = True)
```

```python
#Unique valuelara bax , texmini 10-dan az olanlar kategorical olmalidir
df.nunique()
```

```python
  # Handle with categories
  df['gender'] = df['gender'].astype('category')
  df['smoker'] = df['smoker'].astype('category')
  df['day'] = df['day'].astype('category')
  df['time'] = df['time'].astype('category')
  df['size'] = df['size'].astype('category')
  df.info()
```

```python
#Clear null rows
df_clear = df.dropna()
```

```python
#Check correlation
df_clear.corr()
```

```python
#filter where tip < 6 for outliers

df_clear = df_clear[df_clear['tip'] < 6]
```

```python
##Check corr again
df_clear.corr()
```

```python
# countplot

sns.countplot(x = 'gender', data = df_clear , palette='Reds')
plt.title('Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### Barplot

```python
### Barplot by median
##Check by sum and mean
from numpy import mean
from numpy import median
barplot = sns.barplot(x = "gender" , y = "tip" ,  data = df_clear , palette = 'pastel' , ci = None ,
              estimator = median )
plt.title('Median value by gender')
plt.xlabel('Gender')
plt.ylabel('Median')
plt.show()
```

### Boxplot

```python
##Boxplot
sns.boxplot(y='tip',data=df_clear , color = 'red')
plt.title('Boxplot')
```

```python
##Simple countplot
sns.countplot(x='gender', data=df_clear )
```

```python
#get correlation again
df_clear.corr()
```

```python
## Plot heatmap
tc = df_clear.corr()
sns.heatmap(tc,cmap='coolwarm', annot=True)
plt.title('Correlation heatmap', )
```

```python
#Random heatmap
a = [5,6,7,8,9,9,4,10,12,15]
b =  [ 9,10,20,12,12,17,14,12,10,11]
c = [1,2,3,4,5,3,6,6,87,9]
abc = pd.DataFrame({'a': a, 'b': b, 'c': c})
sns.heatmap(abc.corr(), cmap='RdYlGn', annot=True)
plt.title('Correlation Heatmap')
plt.show()

```

### Piechart

```python
## Value count by gender
df_clear["gender"].value_counts()
```

```python
##Plot pie chart
df_clear["gender"].value_counts().plot.pie(shadow=True, autopct="%1.1f%%")
plt.legend()
plt.title('Pie chart by Region')
```

```python
#round tip and sort

df_clear['tip'] = df_clear['tip'].round(1)
df_clear.sort_values(by='tip', ascending=False)
```
