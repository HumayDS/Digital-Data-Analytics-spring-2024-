```python
##Import packages
from warnings import filterwarnings
filterwarnings('ignore')
import pandas as pd
import statsmodels.api as sm
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
##Import dataset
df = pd.read_csv("https://raw.githubusercontent.com/HumayDS/Big-data-analysis/main/Advertising.csv")
df.head()
```

```python
#Drop first unnecessary column
df.drop("Unnamed: 0" , axis =1 , inplace = True)
```

```python
#Check NA's
df.isnull().sum()
```

```python
#Determine X
X = df[["TV"]]
X[0:5]
```

```python
##Add constant
X = sm.add_constant(X)
X.head()
```

```python
###Determine y
y = df["sales"]
y
```

```python
##Create and fit model
lm = sm.OLS(y,X)
model = lm.fit()
```

```python
model.summary()
```

```python
print("Sales = " +  str("%.2f" % model.params[0]) + " + TV" + "*" + str("%.2f" % model.params[1]))
```

```python
#  calculate rmse of model

# Calculate the root mean squared error (RMSE)
rmse = np.sqrt(mean_squared_error(y, model.predict(X)))

# Print the RMSE
print("RMSE:", rmse)
```

```python
#Create model without sales and newspaper
X = df.drop(["sales",'newspaper'], axis = 1)
y = df["sales"]
X = sm.add_constant(X)
```

```python
lm = sm.OLS(y, X)
```

```python

model = lm.fit()
model.summary()
```

```python
#  calculate rmse of model

# Calculate the root mean squared error (RMSE)
rmse = np.sqrt(mean_squared_error(y, model.predict(X)))

# Print the RMSE
print("RMSE:", rmse)
```

```python
# Split to test and train

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
#  create linear regression and predict on test

# Fit the linear regression model
model = sm.OLS(y_train, X_train).fit()

# Predict on the test set
y_pred = model.predict(X_test)

# Print the predictions
print(y_pred)
```

```python
#  calculate train and  test rmse

# Calculate the root mean squared error (RMSE) for the train set
rmse_train = np.sqrt(mean_squared_error(y_train, model.predict(X_train)))

# Calculate the root mean squared error (RMSE) for the test set
rmse_test = np.sqrt(mean_squared_error(y_test, model.predict(X_test)))

# Print the RMSE values
print("RMSE (train):", rmse_train)
print("RMSE (test):", rmse_test)
```
